Encoding the data. This acts as the LGN which takes the input from the retina to the layers in visual cortex

```python
#DATA ENCODING (ONEHOT) OK

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, OneHotEncoder
import numpy as np

# Load data from CSV file
data = pd.read_csv('/content/Iris.csv')


data.drop('Id', axis=1, inplace=True)

# Modify the target column name and classes
target_column = 'Species'  # Replace with the actual target column name
classes = ['Iris-Setosa', 'Iris-Versicolor']

# Encode the target column
label_encoder = LabelEncoder()
data[target_column] = label_encoder.fit_transform(data[target_column])

# Perform one-hot encoding on the target column
one_hot_encoder = OneHotEncoder(sparse=False, categories='auto')
encoded_target = one_hot_encoder.fit_transform(data[[target_column]])
encoded_feature_names = one_hot_encoder.get_feature_names_out([target_column])
data = pd.concat([data, pd.DataFrame(encoded_target, columns=encoded_feature_names)], axis=1)
data.drop([target_column], axis=1, inplace=True)

# Specify the columns to be normalized
columns_to_normalize = data.columns  # All columns except the one-hot encoded target column

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Apply min-max normalization to the specified columns
data[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])

print(data)

# Split the data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.4, random_state=42)

# To remove the last columns for OHE
features = train_data.iloc[:, :-len(classes)]


train_data_values = features.values.tolist()
data_series = pd.Series(train_data_values)


# Combine all the encoded values into a single bracket
encoded_data = data_series.tolist()


# Normalization
min_range = 20
max_range = 255


# Define the generate_convolution_matrix function to handle a list of values
def generate_convolution_matrix(values, num_neurons):
    element_range = max(-360, min(360, num_neurons))
    step = element_range / (num_neurons - 1) if num_neurons > 1 else 0
    half_range = float(element_range) / 2  # Convert half_range to float
    convolution_matrices = []

    for value in values:
        normalized_value = (value - 0) / (1 - 0) * (max_range - min_range) + min_range
        convolution = [normalized_value + (j * step) - half_range for j in range(num_neurons)]
        convolution_matrices.append(convolution)

    return convolution_matrices

# Get user input for the number of neurons
num_neurons = int(input("Enter the number of neurons: "))

# Initialize a list to store all the convolution matrices
convolution_matrices = []

# Loop through the rows of the normalized data matrix
for value_index in range(len(encoded_data)):
    # Choose a value from the normalized data matrix
    selected_value = encoded_data[value_index]

    # Generate the convolution matrix for the selected value
    convolution_result = generate_convolution_matrix(selected_value, num_neurons)

    # Append the convolution matrix to the list of convolution matrices
    convolution_matrices.append(convolution_result)

# Convert the list of convolution matrices to a NumPy array
convolution_matrices = np.array(convolution_matrices)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Print the resulting convolution matrices
print("Convolution Matrices:")
with np.printoptions(threshold=np.inf, linewidth=np.inf):
    print(convolution_matrices)

```
